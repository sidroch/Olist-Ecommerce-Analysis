# 📊 Análises e Gráficos — Relatório Executivo Olist

Este notebook contém os gráficos utilizados no Relatório Executivo do Tech Challenge Olist.
Os dados são carregados a partir dos arquivos CSV localizados na pasta `../data`.

---

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
sns.set(style="whitegrid", palette="muted", font_scale=1.1)
```

## 📥 Carregamento dos Dados
Os arquivos foram extraídos do PPT e salvos em `../data`.

```python
df_logistica = pd.read_csv("../data/logistica.csv")
df_cat = pd.read_csv("../data/categorias.csv")
df_receita = pd.read_csv("../data/receita_estado.csv")
df_riscos = pd.read_csv("../data/riscos_oportunidades.csv")

df_logistica.head(), df_cat.head(), df_receita.head(), df_riscos.head()
```

## 1️⃣ Percentual de Atrasos por Estado
Gráfico de barras ordenado pelo percentual de atrasos.

```python
df_logistica_sorted = df_logistica.sort_values("percentual_atrasos", ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(data=df_logistica_sorted, x="estado", y="percentual_atrasos", color="tomato")
plt.title("Percentual de Atrasos por Estado")
plt.xlabel("Estado")
plt.ylabel("% de Atrasos")
plt.show()
```

## 2️⃣ Destaques por Categoria
Ranking simples baseado nos destaques qualitativos do relatório.

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=df_cat["categoria"], y=list(range(len(df_cat))), palette="viridis")
plt.title("Destaques por Categoria")
plt.xlabel("Categoria")
plt.ylabel("Ranking (1 = maior destaque)")
plt.show()
```

## 3️⃣ Receita por Estado
Ranking representando os insights de receita por estado.

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=df_receita["estado"], y=list(range(len(df_receita))), color="purple")
plt.title("Insights de Receita por Estado")
plt.xlabel("Estado")
plt.ylabel("Ranking de Receita / Insight")
plt.show()
```

## 4️⃣ Riscos e Oportunidades por Estado
Classificação visual dos estados conforme o relatório.

```python
cores = df_riscos["tipo"].map({"Risco": "red", "Oportunidade": "green"})

plt.figure(figsize=(10, 6))
sns.barplot(x=df_riscos["estado"], y=list(range(len(df_riscos))), palette=cores)
plt.title("Riscos e Oportunidades por Estado")
plt.xlabel("Estado")
plt.ylabel("Classificação")
plt.show()
```

## 5️⃣ RJ vs Mediana — Percentual de Atrasos
Comparação direta entre o percentual de atrasos do RJ e a mediana dos estados.

```python
rj = 12
mediana = 9.29

plt.figure(figsize=(6, 4))
sns.barplot(x=["RJ", "Mediana"], y=[rj, mediana], palette=["red", "gray"])
plt.title("RJ vs Mediana — Percentual de Atrasos")
plt.ylabel("% de Atrasos")
plt.show()
```

## ✅ Conclusão Executiva

Os gráficos apresentados reforçam os principais achados do relatório, evidenciando padrões claros de desempenho comercial e eficiência logística entre os estados analisados. As diferenças regionais destacam oportunidades estratégicas específicas, enquanto os indicadores de atraso demonstram o impacto direto da operação logística na experiência do cliente e na rentabilidade. Esses resultados fortalecem a importância de decisões orientadas por dados para impulsionar crescimento sustentável e competitividade no e‑commerce.
