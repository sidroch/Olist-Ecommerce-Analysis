# Gráficos do Relatório Executivo Olist

Este notebook reproduz, em Python, os principais gráficos apresentados no PPT do Tech Challenge Olist.

Os dados são carregados a partir dos arquivos CSV localizados na pasta `../data`.

```python
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

## Carregamento dos CSVs

Os arquivos foram extraídos do PPT e salvos em `../data`.

```python
df_logistica = pd.read_csv("../data/logistica.csv")
df_cat = pd.read_csv("../data/categorias.csv")
df_receita = pd.read_csv("../data/receita_estado.csv")
df_riscos = pd.read_csv("../data/riscos_oportunidades.csv")
```

## 1. Percentual de Atrasos por Estado

Gráfico de barras ordenado pelo percentual de atrasos.

```python
df_logistica_sorted = df_logistica.sort_values("percentual_atrasos", ascending=False)

plt.figure(figsize=(8, 5))
plt.bar(df_logistica_sorted["estado"], df_logistica_sorted["percentual_atrasos"], color="tomato")
plt.title("Percentual de Atrasos por Estado")
plt.xlabel("Estado")
plt.ylabel("% de Atrasos")
plt.grid(axis="y", linestyle="--", alpha=0.4)
plt.show()
```

## 2. Destaques por Categoria (Volume, Receita, Valor Agregado)

O PPT traz destaques qualitativos, então aqui usamos um ranking simples.

```python
plt.figure(figsize=(8, 5))
plt.bar(df_cat["categoria"], range(len(df_cat)), color=["steelblue", "seagreen", "gold"])
plt.title("Destaques por Categoria")
plt.xlabel("Categoria")
plt.ylabel("Ranking (1 = maior destaque)")
plt.show()
```

## 3. Receita por Estado (Insights do PPT)

Representação dos insights de receita por estado em forma de ranking.

```python
plt.figure(figsize=(8, 5))
plt.bar(df_receita["estado"], range(len(df_receita)), color="purple")
plt.title("Insights de Receita por Estado")
plt.xlabel("Estado")
plt.ylabel("Ranking de Receita / Insight")
plt.show()
```

## 4. Riscos e Oportunidades por Estado

Estados classificados como risco (vermelho) ou oportunidade (verde).

```python
cores = df_riscos["tipo"].map({"Risco": "red", "Oportunidade": "green"})

plt.figure(figsize=(8, 5))
plt.bar(df_riscos["estado"], range(len(df_riscos)), color=cores)
plt.title("Riscos e Oportunidades por Estado")
plt.xlabel("Estado")
plt.ylabel("Classificação")
plt.show()
```

## 5. RJ vs Mediana — Percentual de Atrasos

Comparação direta entre o percentual de atrasos do RJ e a mediana dos estados.

```python
rj = 12
mediana = 9.29

plt.figure(figsize=(6, 4))
plt.bar(["RJ", "Mediana"], [rj, mediana], color=["red", "gray"])
plt.title("RJ vs Mediana — Percentual de Atrasos")
plt.ylabel("% de Atrasos")
plt.show()
```
